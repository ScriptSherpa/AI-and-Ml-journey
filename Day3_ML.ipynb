```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
a = df[['Age', 'EstimatedSalary']]
a
```

```python
plt.figure(figsize=(16,9))

plt.title('Data Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Age')
plt.scatter(a['EstimatedSalary'], a['Age'], color = 'red');
plt.grid()
```

```python
# Import the class

from sklearn.preprocessing import MinMaxScaler
```

```python
# Create an Object
scaler = MinMaxScaler()
```

```python
a_scaled = scaler.fit_transform(a)
```

```python
a_scaled;
```

```python
a_scaled[:,0].min(), a_scaled[:,0].max()
```

```python
a_scaled[:,1].min(), a_scaled[:,1].max()
```

```python
pd.DataFrame(a_scaled).describe()
```

```python
plt.figure(figsize=(16,9))

plt.title('Data Distribution in MinMaxScaler')
plt.xlabel('Estimated Salary')
plt.ylabel('Age')
plt.scatter(a_scaled[:,1], a_scaled[:,0], color = 'blue');
plt.grid()
```

```python
# Import the class

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
a_scaled = scaler.fit_transform(a)
```

```python
a_scaled;
```

```python
plt.figure(figsize=(16,9))

plt.title('Data Distribution using StandardScaler')
plt.xlabel('Estimated Salary')
plt.ylabel('Age')
plt.scatter(a_scaled[:,1], a_scaled[:,0], color = 'orange');
plt.grid()
```
